```python
import pandas as pd
```

```python
data = pd.read_csv('../data/wmt15.en-cs.csv', sep='\0', header=None, names=['en'])
data['en'] = data['en'].str.replace(r"[^\w '\.,\-]", "", regex=True)
data = data[data['en_len'] >= 10]
```

```python
data['en_len']  = data['en'].str.len()
data[data['en_len'] > 2].head()
data = data.sort_values(by='en_len')
data['cumsum'] = data['en_len'].cumsum()
```

```python
selected_data = data[data['cumsum'] < 2_000_000]
```
