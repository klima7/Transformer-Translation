```python
import pandas
import numpy as np
```

```python
data = pandas.read_csv('../data/en-fr.csv')
```

```python
data = data.dropna()
```

```python
data['en_len']  = data['en'].str.len()
data['fr_len']  = data['fr'].str.len()

data = data.sort_values(by='en_len')
```

```python
data['tmp'] = data['en'].str.replace(r"[^\w '\.,\-]", "", regex=True)
```

```python
min_length = 20
lenghten_factor = 0.4

data = data[
    (data['en_len'] >= min_length) & 
    (data['fr_len'] >= min_length) & 
    (abs(data['en_len']-data['fr_len'])<data['en_len']*lenghten_factor)
]
```

```python
data['cumsum'] = data['en_len'].cumsum()
```

```python
selected_data = data[data['cumsum'] < 2_000_000]
```
